```python
# Taking three integer values from the user
a = int(input("Enter the first integer value: "))
b = int(input("Enter the second integer value: "))
c = int(input("Enter the third integer value: "))

# Finding the median
if (a >= b and a <= c) or (a <= b and a >= c):
    median = a
elif (b >= a and b <= c) or (b <= a and b >= c):
    median = b
else:
    median = c

# Printing the median
print("The median of the three integers is:", median)
  
```

```python
marks = int(input("Enter your marks: "))
if marks < 25:
    print("Your grade is F.")
elif marks >= 25 and marks <= 44:
    print("Your grade is E.")
elif marks >= 45 and marks <= 49:
    print("Your grade is D.")
elif marks >= 50 and marks <= 59:
    print("Your grade is C.")
elif marks >= 60 and marks <= 79:
    print("Your grade is B.")
elif marks >= 80 and marks <= 89:
    print("Your grade is A.")
else:
    print("Your grade is A+.")
```
